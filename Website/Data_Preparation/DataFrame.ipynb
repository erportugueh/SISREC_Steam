```python
# Step 1: Get top 30 most common genres (including NaN initially)
top_30_genres = df["Genres"].value_counts(dropna=False).head(30)

# Step 2: Remove NaN
top_30_genres_no_nan = top_30_genres[~top_30_genres.index.isna()].index

# Step 3: Create an empty list to store top games per genre
top_games_per_genre = []

# Step 4: Loop through each genre and get top 30 by Estimated owners
for genre in top_30_genres_no_nan:
    genre_df = df[df["Genres"] == genre]
    
    # Make sure 'Estimated owners' is numeric (if not already)
    genre_df = genre_df.copy()
    genre_df["Estimated owners"] = pd.to_numeric(genre_df["Estimated owners"], errors="coerce")
    
    # Drop rows with NaN in Estimated owners
    genre_df = genre_df.dropna(subset=["Estimated owners"])

    # Sort by Estimated owners descending, take top 30
    top_30 = genre_df.sort_values(by="Estimated owners", ascending=False).head(30)
    
    top_games_per_genre.append(top_30)

# Step 5: Combine all top games into one DataFrame
filtered_df = pd.concat(top_games_per_genre, ignore_index=True)

# Step 6: Save to CSV
filtered_df.to_csv("top_30_genres_top_30_by_owners.csv", index=False)

print("‚úÖ Saved filtered DataFrame with top 30 games per genre (by Estimated owners) to 'top_30_genres_top_30_by_owners.csv'")

```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load the dataset
df = pd.read_csv("top_30_genres_top_30_by_owners.csv")

# Drop NaNs and get unique genres
unique_genres = df['Genres'].dropna().unique()

# Vectorize the genre strings
vectorizer = TfidfVectorizer()
genre_vectors = vectorizer.fit_transform(unique_genres)

# Compute cosine similarity between genres
similarity_matrix = cosine_similarity(genre_vectors)

# Mask to ignore self-comparisons and duplicates
mask = np.triu(np.ones(similarity_matrix.shape), k=1).astype(bool)
flattened_similarities = similarity_matrix[mask]

# Get top 20 most similar genre pairs
top_indices = np.argsort(flattened_similarities)[-20:]
rows, cols = np.where(mask)
top_pairs = list(zip(rows[top_indices], cols[top_indices], flattened_similarities[top_indices]))

# Create a readable DataFrame
similar_genres_df = pd.DataFrame(
    [(unique_genres[i], unique_genres[j], sim) for i, j, sim in top_pairs],
    columns=["Genre 1", "Genre 2", "Similarity"]
).sort_values(by="Similarity", ascending=False)

# Display results
print("üéØ Top 20 Most Similar Genre Pairs:")
print(similar_genres_df.to_string(index=False))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# --------------------------
# Load the dataset
# --------------------------
df = pd.read_csv("top_30_genres_top_30_by_owners.csv")

# --------------------------
# User input: game names
# --------------------------
user_games = [
    "Hollow Knight",
    "Cuphead",
    "The Jackbox Party Pack 9",
    "Fishing The Abyss"
]

# Normalize input
user_games_clean = [g.lower().strip() for g in user_games]
df['Name_clean'] = df['AppID'].str.lower().str.strip()

# Filter only selected games
selected_games = df[df['Name_clean'].isin(user_games_clean)]

if selected_games.empty:
    print("üö´ None of the input games were found. Check the names.")
else:
    # Extract their genre strings and drop duplicates
    genres = selected_games['Genres'].dropna().unique()

    if len(genres) < 2:
        print("üö´ Not enough distinct genres to compare.")
    else:
        # Vectorize genre strings
        vectorizer = TfidfVectorizer()
        genre_vectors = vectorizer.fit_transform(genres)

        # Compute cosine similarity
        similarity_matrix = cosine_similarity(genre_vectors)

        # Mask for upper triangle (no self-comparisons)
        mask = np.triu(np.ones(similarity_matrix.shape), k=1).astype(bool)
        similarities = similarity_matrix[mask]

        # Get top genre pairs
        rows, cols = np.where(mask)
        pairs = list(zip(rows, cols, similarities))

        # Sort and keep top 10
        top_pairs = sorted(pairs, key=lambda x: x[2], reverse=True)[:10]

        # Show results
        print("üéØ Top Genre Similarities From Selected Games:\n")
        for i, j, sim in top_pairs:
            print(f"{genres[i]}  ‚áÑ  {genres[j]}   ‚Üí Similarity: {sim:.3f}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# --------------------------
# Load the dataset
# --------------------------
df = pd.read_csv("top_30_genres_top_30_by_owners.csv")

# --------------------------
# User input: game names
# --------------------------
user_games = [
    "Hollow Knight",
    "Cuphead",
    "The Jackbox Party Pack 9",
    "Hollow Cocoon",
    "Fishing The Abyss"
]

# Normalize input
user_games_clean = [g.lower().strip() for g in user_games]
df['Name_clean'] = df['AppID'].str.lower().str.strip()

# Filter only selected games
selected_games = df[df['Name_clean'].isin(user_games_clean)]

if selected_games.empty:
    print("üö´ None of the input games were found. Check the names.")
else:
    # Step 1: Get unique genre strings from all games
    df['Genres'] = df['Genres'].fillna('')
    all_genres = df['Genres'].unique()

    # Step 2: Vectorize all genres
    vectorizer = TfidfVectorizer()
    all_genre_vectors = vectorizer.fit_transform(all_genres)

    # Step 3: Vectorize user game genres
    user_genres = selected_games['Genres'].dropna().unique()
    user_genre_vectors = vectorizer.transform(user_genres)

    # Step 4: Compute cosine similarity between each user genre and all genres
    sim_matrix = cosine_similarity(user_genre_vectors, all_genre_vectors)

    # Step 5: Aggregate similarity (max similarity across user genres)
    max_similarities = sim_matrix.max(axis=0)

    # Step 6: Get top 10 most similar genre strings
    top_indices = max_similarities.argsort()[::-1][:10]

    print("\nüåü Top 10 Genres Similar to User's Game Genres:\n")
    for i in top_indices:
        genre_str = all_genres[i]
        similarity = max_similarities[i]
        print(f"{genre_str} ‚Üí Similarity Score: {similarity:.3f}")
```
